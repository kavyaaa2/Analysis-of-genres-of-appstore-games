## <b>â¢Perform Data Preparation by cleaning the data and removing null values </b>

Approach:

-To fill in the NaN values in the columns of the dataframe which are not necessary to analyze the app.

-Fill in the null values of string type as "-".

-Fill in the NaN values of price and in-app purchases as 0.

-We are going to leave the NaN values of User ratings and Average users as it is, so that it can be dropped later as calculations for those particular apps are not possible

```python
df["URL"].fillna("Not available",inplace=True)
df["ID"].fillna("-",inplace=True)
df["Subtitle"].fillna("Not available",inplace=True)
df["Icon URL"].fillna("-",inplace=True)
df["Price"].fillna(0,inplace=True)
df["In-app Purchases"].fillna(0,inplace=True)
df["Description"].fillna("-",inplace=True)
df["Developer"].fillna("-",inplace=True)
df["Age Rating"].fillna("-",inplace=True)
df["Languages"].fillna("EN",inplace=True)
df["Primary Genre"].fillna("Games",inplace=True)
df["Genres"].fillna("-",inplace=True)
df["Original Release Date"].fillna("-",inplace=True)
df["Current Version Release Date"].fillna("-",inplace=True)
```

-Drop the NaN values for further possible calculations.

```python
df=df.dropna()
df
```

```python
l1=list(df['Genres'])
l1
```

-Split the strings in list and check for redundancy.

```python
l2=[]
l3=[]
count=0
for a in l1:
    l3=a.split(',')
    for b in l3:
        if b.strip() not in l2:
            count+=1
            l2.append(b.strip())
print(l2)
print(count)
len(l2)
```

-We'll calculate the average rating, total number of users,total apps and average price per genre.

```python
avg_genre=[]  #average rating per genre
user_genre=[] #number of users using apps of a particular genre
avg_price=[]  #average price per genre
apps_genre=[] #total apps per genre
for a in range(0,39):
    p=dict_genre[a]["Average User Rating"].sum()
    q=dict_genre[a].shape[0]
    avg=p/q
    avg_genre.append(avg)
    user=dict_genre[a]["User Rating Count"].sum()
    user_genre.append(user)
    r=dict_genre[a]["Price"].sum()
    s=r/q
    avg_price.append(s)
    apps_genre.append(q)
avg_genre
```

-Now we'll we catagorize the genres in 2 catagories namely:

1.High rated genres<i> (all the genres with average rating 4.0 and above)</i>

2.Low rated genres<i> (all the genres with average rating beolw 4.0)</i>

-Also, calculate average rating, total number of users,total apps and average price per genre seperately for above mentioned genres for further analysis.

```python
h_genre=[]      #genres having ratings above 4.0
l_genre=[]      #genres having ratings below 4.0
h_avg_genre=[]  #High average rating genres
l_avg_genre=[]  #low average rating genres
h_user_genre=[] #High average rating genres no. of users
l_user_genre=[] #Low average rating genres no. of users
h_avg_price=[]  #Average price in high rated genres
l_avg_price=[]  #Average price in high rated genres

for a in range(0,39):
    if avg_genre[a]>=4.0:
        h_genre.append(l2[a])
        h_avg_genre.append(avg_genre[a])
        h_user_genre.append(user_genre[a])
        h_avg_price.append(avg_price[a])
    else:
        l_genre.append(l2[a])
        l_avg_genre.append(avg_genre[a])
        l_user_genre.append(user_genre[a])
        l_avg_price.append(avg_price[a])

print("HIGH rated genres are:\n")
print(h_genre)
print("\n\nLOW rated genres are:\n")
print(l_genre)
```
